```python
import requests
import os
import csv
from typing import List
from bs4 import BeautifulSoup
```

```python
# do your magic
link_to_scrape = 'https://www.cards-capital.com/1041-cartes-pouvoir-des-elements-pote-a-l-unite'
data = requests.get(link_to_scrape)
csv_data = []

#by @Ivan vinogradov
def download(urls: List[str], dest_folder: str):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)  # create folder if it does not exist
    for url in urls:
        filename = url.split('/')[-1].replace(" ", "_")  # be careful with file names
        file_path = os.path.join(dest_folder, filename)

        r = requests.get(url, stream=True)
        if r.ok:
            print("saving to", os.path.abspath(file_path))
            with open(file_path, 'wb') as f:
                for chunk in r.iter_content(chunk_size=1024 * 8):
                    if chunk:
                        f.write(chunk)
                        f.flush()
                        os.fsync(f.fileno())
        else:  # HTTP status code 4XX/5XX
            print("Download failed: status code {}\n{}".format(r.status_code, r.text))
############################################################################################################################

if data.status_code == 200:
    soup = BeautifulSoup(data._content)
    label_divs = soup.find_all('strong',{"class":"edition"})
    label_divs = [x.contents[0] for x in label_divs]
    card_names = soup.find_all('strong',{"class":"name"})
    card_names = [x.contents[0] for x in card_names]
    card_a = soup.find_all('a',{'class':"zoomhover"})
    card_links=[]
    for link in card_a:
        card_image_link = BeautifulSoup((requests.get(link.get('href')))._content).find(id='productimagelink').get('href')
        card_links.append(card_image_link)
```
